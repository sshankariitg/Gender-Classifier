```python
#Cropping all the images

def img_extractor(address,gender,i):
    img=cv2.imread(address)
    gray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
    faces=haar.detectMultiScale(gray,1.5,5)
    for x,y,w,h in faces:
        detect=img[y:y+h,x:x+w]
        if gender=='male':
              cv2.imwrite('./data/cropped_images/male/{}_{}.png'.format(gender,i),detect)
        else:
            cv2.imwrite('./data/cropped_images/female/{}_{}.png'.format(gender,i),detect)
              
        
        
```

```python
for i,address in enumerate(female):
    try:
        img_extractor(address,'female',i)
        print('INFO:{}/{} PROCESSED SUCCESSFULLY'.format(i,len(female)))
    except:
        print('INFO:{}/{} SORRY CANNOT PROCESS'.format(i,len(female)))
```

```python
for i,address in enumerate(male):
    try:
        img_extractor(address,'male',i)
        print('INFO:{}/{} PROCESSED SUCCESSFULLY'.format(i,len(male)))
    except:
        print('INFO:{}/{} SORRY CANNOT PROCESS'.format(i,len(male)))
```
